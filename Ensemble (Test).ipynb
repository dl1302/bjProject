```python
from pathlib import Path

#file = Path.joinpath('Resources', 'test.txt')
file = Path.cwd() / r"C:\Users\danny\helloWorld\Energies3\44_A.txt"
#file.read_text()

    # Open the file in "read" mode ('r') 
with open(file, 'r') as text:
    textfile = text.read()
    print(textfile)
```

```python
rawData = textfile.split()
for i in range(len(rawData)):
    rawData[i] = round(float(rawData[i]),2)
print(rawData)
```

```python
del rawData[0]
print(rawData)
```

```python
import math
```

```python
x = rawData[len(rawData)-1]
E = []
A = []
lim = rawData[0]
lim = lim+.01
while (x<lim):
    E.append(x)
    Sum = 0
    for i in range(len(rawData)):
        a = x - rawData[i]
        b = a*a
        c = -b/.005
        e = math.exp(c)
        Sum = Sum+e
    A.append(Sum)
    x = x+.01
    x = round(x,2)
print(E)
print(A)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(E,A)
plt.xlim([290.7,291.4])
plt.ylim([0,20])
plt.show()
```
