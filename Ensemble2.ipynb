```python
Energies = []
```

```python
for i in range(70):
    f = str(i) + '_A.txt'
    fd = open(f,'r')
    textfile = fd.read()
    rawData = textfile.split()
    for j in range(len(rawData)):
        rawData[j] = round(float(rawData[j]),2)
    toRemove = []
    for j in range(len(rawData)):
        if (rawData[j]>310):
            toRemove.append(rawData[j])
    for j in range(len(toRemove)):
        rawData.remove(toRemove[j])
    for j in range(len(rawData)):
        Energies.append(rawData[j])
print(Energies)
```

```python
print(len(Energies))
```

```python
import math
```

```python
x = 280
E = []
A = []
while (x<310.01):
    E.append(x)
    Sum = 0
    for i in range(len(Energies)):
        a = x - Energies[i]
        b = a*a
        c = -b/.001
        e = math.exp(c)
        Sum = Sum+e
    A.append(Sum)
    x = x+.01
    x = round(x,2)
```

```python
for i in range(len(E)):
    E[i] = E[i]-6.2
    E[i] = round(E[i],2)
for i in range(len(A)):
    A[i] = 3.3*A[i]/40.4
print(E)
print(A)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(E,A)
plt.xlim([286,284])
plt.ylim([0,30])
plt.xlabel('Ionization Energy eV')
plt.ylabel('Relative Absorbance')
plt.legend(["Experimental"],loc="upper right")
plt.show()
```
