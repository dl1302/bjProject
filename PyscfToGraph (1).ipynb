```python
from pathlib import Path

#file = Path.joinpath('Resources', 'test.txt')
file = Path.cwd() / r"C:\Users\danny\helloWorld\Energies3\44_A.txt"
#file.read_text()

    # Open the file in "read" mode ('r') 
with open(file, 'r') as text:
    textfile = text.read()
    print(textfile)
```

```python
rawData = textfile.split()
for i in range(len(rawData)):
    rawData[i] = float(rawData[i])
print(rawData)
```

```python
data = []
length = len(rawData)
for i in range(length):
    if i == 0:
        data.append([rawData[i]])
    else:
        check = 0
        for j in range(i):
            if rawData[i]==rawData[j]:
                check = check+1
        if check==0:
            data.append([rawData[i]])
print(data)
```

```python
y = []
for i in range(len(data)):
    y.append(3*rawData.count(data[i][0]))
print(y)
```

```python
x = []
for i in range(len(data)):
    x.append(data[i][0])
print(x)
```

```python
forExcel = []
forExcel.append(y)
forExcel.append(x)
print(forExcel)
```
