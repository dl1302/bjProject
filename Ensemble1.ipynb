```python
for i in range(70):
    if (i!=17) and (i!=19) and (i!=32) and (i!=37):
        f = str(i) + '_A.txt'
        fd = open(f,'r')
        textfile = fd.read()
        rawData = textfile.split()
        for j in range(len(rawData)):
            rawData[j] = round(float(rawData[j]),2)
        toRemove = []
        for j in range(len(rawData)):
            if (rawData[j]>310):
                toRemove.append(rawData[j])
        for j in range(len(toRemove)):
            rawData.remove(toRemove[j])
        x = 280
        E = []
        A = []
        while (x<310.01):
            E.append(x)
            Sum = 0
            for i in range(len(rawData)):
                a = x - rawData[i]
                b = a*a
                c = -b/.002
                e = math.exp(c)
                Sum = Sum+e
            A.append(Sum)
            x = x+.01
            x = round(x,2)
        for j in range(len(bigA)):
            bigA[j] = bigA[j] + A[j]
```

```python
print(bigA)
```

```python
plt.scatter(bigE,bigA)
plt.xlim([290,293])
plt.ylim([0,325])
plt.show()
```
