```python
f = 'B60.txt'
fd = open(f,'r')
textfile = fd.read()
rawData = textfile.split()
print(rawData)
```

```python
for j in range(len(rawData)):
    rawData[j] = round(float(rawData[j]),2)
print(rawData)
```

```python
import math
```

```python
Energies = rawData
x = 280
E = []
A = []
while (x<310.01):
    E.append(x)
    Sum = 0
    for i in range(len(Energies)):
        a = x - Energies[i]
        b = a*a
        c = -b/4
        e = math.exp(c)
        Sum = Sum+e
    A.append(Sum)
    x = x+.01
    x = round(x,2)
print(E)
print(A)
```

```python
for i in range(len(A)):
    A[i] = A[i]*1000
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(E,A)
plt.xlim([310,280])
plt.ylim([0,10])
plt.show()
```
